```python
import pandas as pd
import numpy as np
```

**Importing data into a dataframe**

```python
df = pd.read_csv('../input/snapshot-round-1/snapshot_round_1.csv')
df.head(10)
```

**Replacing blank fields(NaN) in the campaign colum with “Offer_NoApp_postnatal”**

```python
df['campaign_name'] = df['campaign_name'].replace(np.NaN, 'Offer_NoApp_postnatal')
df.head(10)
```

**Data Cleaning**

```python
df['campaign_name'].unique()
```

```python
df['campaign_name'] = df['campaign_name'].replace(['Newsletter_postnatal_NohApp', 'Hello_Email_postnatall_NoApp'], ['Newsletter_postnatal_NoApp', 'Hello_Email_postnatal_NoApp'])
df.head(10)
```

```python
# Adding Age and HasApp columns

df['Age'] = np.where(df['campaign_name'].str.contains('prenatal'), 'prenatal', 'postnatal')
df['HasApp'] = np.where(df['campaign_name'].str.contains('WithApp'), 'True', 'False')
df.head(10)
```

```python
# renaming columns
df.rename(columns = {'mktng_pgm_nbr':'Country_code', 'campaign_name': 'Campaign_name'}, inplace = True)

# reindexing columns
df=df.reindex(columns=['date', 'Country_code', 'Campaign_name', 'Age', 'HasApp', 'sendout', 'delivered', 'people', 'unique_delivered', 'openers', 'clickers', 'clicked_or_opened', 'all_opens', 'all_clicks'])
df.head()
```

```python
df['date'] = pd.to_datetime(df['date'])
df.groupby(pd.Grouper(key='date', axis=0, 
                      freq='M'), ).sum()
```

```python
to_drop = ['sendout',
           'delivered',
           'people',
           'unique_delivered',
           'openers',
           'clickers',
           'clicked_or_opened',
           'all_opens',
           'all_clicks']

df.drop(to_drop, inplace = True, axis = 1)
df.head()
```
